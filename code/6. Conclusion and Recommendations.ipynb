## Model and Error Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import math
import datetime as dt
import warnings
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE 
from sklearn.metrics import (confusion_matrix, classification_report, plot_roc_curve, roc_auc_score, 
accuracy_score, precision_score, recall_score, f1_score, auc, precision_recall_curve, average_precision_score)

%config InlineBackend.figure_format = 'retina'
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
```

```python
#Importing model and datasets for analysis

final_lr_model = pickle.load(open('./models/final_lr_model.sav', 'rb'))

final_train_df = pd.read_csv('../data/final_train.csv')

weather_train_df = pd.read_csv('../data/weather_train_merged.csv')

model_test_data = pd.read_csv('../data/model_test_data.csv')
```

```python
# fitting model to get predictions for model and error analysis

x = final_train_df.drop('wnvpresent',axis=1)
y = final_train_df['wnvpresent']

display(y.value_counts(normalize=True))

x_train, x_test, y_train, y_test= train_test_split(x,y,random_state=42)

ss = StandardScaler()

x_train_sc = ss.fit_transform(x_train)

x_test_sc = ss.transform(x_test)

sm = SMOTE(random_state=42)

x_train_sc, y_train = sm.fit_resample(x_train_sc,y_train)

final_lr_model.fit(x_train_sc,y_train)

predictions = final_lr_model.predict(x_test_sc)
```

```python
# Filtering out top coefficients for wnv predictions

log_coef = pd.DataFrame(final_lr_model.best_estimator_.steps[0][1].coef_.T,x_test.columns,
                        columns = ['coefficients']).sort_values(by='coefficients', ascending=False)

log_coef['expo_coef'] = log_coef['coefficients'].apply(np.exp)

display(log_coef.head(10))

log_coef.head(10).sort_values(by = 'expo_coef',ascending=True)['expo_coef'].plot(kind='barh')

plt.title('Top 10 Correlated Features')
```

We can interprete the exponentiated coefficients, for example like September (month 8), as being 2 times more likely to classify the instance as having WNV when it is in September. 

We can see that dewpoint, temperature, number of mosquitos, time and traps are strong predictors of WNV. Hence, we will inspect our misclassifications with these features to see if they are inline with what we expect.

```python
# Filtering out weekday counts for july-september

weather_train_df[(weather_train_df['month']==7)|\
                 (weather_train_df['month']==8)|\
                 (weather_train_df['month']==9)]['weekday'].value_counts()
```

We can see that for the months of Jul-Sep where there is high WNV, there were much more samples collected from wednesday to friday. Hence, it is natural that wed-fri were seen as good predictors for WNV as there were simply more cases collected. It is thus recommended to either ensure that collection frequency is equal on all days, or to drop this feature from our model.

```python
# creating a dataframe for error analysis

error_analysis_df = pd.DataFrame(y_test)
error_analysis_df['predictions'] = final_lr_model.predict(x_test_sc)
error_analysis_df.rename(columns= {'wnvpresent':'actual'},inplace=True)
```

Having a false negative is much more serious than a false positive, as a false negative would mean that we did not detect the presence of WNV and hence fall short of possible ramification steps. This may result in serious healthcare issues from coma, miscarriages or even loss of life <sup> 1 </sup>. Hence we would want to minimize false positives as much as possible at the expense of higher false positives, though that may result in an increase in government expenditure in terms of prevention. We will thus try to understand our false negative misclassifications better.

## Cost Benefit Analysis

We will now explore a reported study to get the estimated cost benefit analysis figures for spraying of insecticides to prevent WNV infections <sup> 2 </sup>. In 2005, there was an outbreak of West Nile Virus (WNV) disease in Sacramento County, California, with 163 human cases reported. The Sacramento-Yolo Mosquito and Vector Control District conducted an emergency aerial spray. The WNV disease in Sacramento County cost ≈ \\$2.28 million for medical treatment and patients’ productivity loss for both West Nile fever (WNF) and West Nile neuroinvasive disease (WNND), which is a type of infection from WNV. Vector control cost ≈ $701,790, including spray procedures and overtime hours. The total economic impact of WNV was \\$2.98 million.</sup>

Out of the 163 cases, there were 46 WNND cases and 117 WNF cases. Total WNND medical and productivity cost is $2,625,200, while the total cost for WNF cases is \\$136,900. WNND cases are much more serious and requires longer inpatient care <sup> 1</sup>, and hence have a higher cost of \\$57k per case, as compared to WNF of \\$1.2k per case.

```python
# Calculating total cost per case

total_cases_2005 = 163
total_cost_2005 = 2625200 + 136900
cost_per_case = round((total_cost_2005/total_cases_2005),-3)
print(f'The cost per case is an estimated ${cost_per_case:,}')
```

Two researchers, Elnaiem et al. and Carney et al., examined the efficacy of the 2005 emergency aerial spray in Sacramento County, which used pyrethrins as the active ingredients to control adult mosquitoes. In both studies, an unsprayed area within the county was used as the control. They found that there was a decrease of 57.5% WNV-competent vector mosquitoes for the sprayed areas against the unsprayed areas.

### Emergency Vector Control Spray

The emergency spray comprised 1,157 additional overtime hours in SYMVCD for August 2005. These overtime hours cost ≈ \\$41,790. The emergency spray cost ≈ \\$660,000, with the total emergency aerial spray response to the WNV epidemic costing a total of $701,790. The spray uses aerial ultra-low–volume adulticiding over 2 areas in Sacramento County comprising ≈ 477 km2. Aerial spraying was conducted on 6 nights in early and mid-August. 

With these figures, we can claculate the estimated costs of spraying. Please see below for the spray areas which we have identified.
```python
#Calculate total cost per km^2

total_cost_2005 = 701790
num_spray_2005 = 6 
km_spray_2005 = 477

spraycost_per_km = total_cost_2005/num_spray_2005/km_spray_2005
round(spraycost_per_km,2)
```

```python
def get_distance(lat1, lon1, lat2, lon2): 
    
    ''' Function to calculate kilometers from total grid size'''
    
    R = 6373.0
    
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    dlon = lon2 - lon1 
    dlat = lat2 - lat1

    temp = (  
         math.sin(dlat / 2) ** 2 
       + math.cos(lat1) 
       * math.cos(lat2) 
       * math.sin(dlon / 2) ** 2
    )
    
    C = (2 * math.atan2(math.sqrt(temp), math.sqrt(1 - temp)))

    distance = R * C
    
    return distance
```

```python
#calculating total grid size

long = get_distance(41.6,-88, 42.1,-88)
latitude = get_distance(41.6,-88, 41.6,-87.5)
total_area = round(long*latitude)

# cut our total grid into 50 blocks to calculate cost per block

area_per_block = total_area/50
area_per_block
```

```python
#calculating total cost per block

total_cost_per_block = area_per_block*spraycost_per_km
total_cost_per_block

#calculating total cost

spray_blocks = 21

spray_freq = 12

total_spray_cost = round(total_cost_per_block * spray_blocks * spray_freq)

total_spray_cost
```

Now that we have our total cost, we will have to compare against the benefits that can be obtained from spraying. To do this, we will identify the incidences of west nile virus from our identified spray locations from the year of our latest data (2013).

```python
# Identifying incidences of west nile virus

top_traps_to_spray = top_traps_df[(top_traps_df['year']==2013) & ((top_traps_df['month']==7)|(top_traps_df['month']==8)|(top_traps_df['month']==9))]

top_traps_to_spray['wnvpresent'].value_counts()
```

As it has proven that spraying can reduce WNV presence by 57.5%, we can estimate the decrease in WNV cases through spraying those areas.

```python
# Calculating the estimated number of WNV incidences reduction after spray
estimated_wnv_reduction = top_traps_to_spray['wnvpresent'].value_counts()[1]*0.575
estimated_wnv_reduction
```

With the estimated figures of reduced WNV after spraying, we can now calculate the benefits of spraying. 

```python
# Estimating reduced cost
# Assuming number of humans infected per wnv mosquito case is 3 

human_cases_per_wnv_case = 3

total_savings = estimated_wnv_reduction * cost_per_case * human_cases_per_wnv_case

total_cost_savings = total_savings- total_spray_cost

print(f'Cost savings assuming 3 human infections per WNV incident detected: ${total_cost_savings:,}')

# Estimated cost savings per additional case of human infection if exceed 3 infections for all incidences

total_cost_savings_per_add_case = estimated_wnv_reduction * cost_per_case

print(f'Cost savings per additional human infection per WNV incident detected: ${total_cost_savings_per_add_case:,}')
```

We can see that if each case of WNV incident infects as little as 3 people, the city would have achieved cost savings of $863k. For each additional human infection per WNV incidence, the city would save an additional \\$1,104,575. That is to say if each WNV incidence would cause 4 human infections, the city would save an additional \\$1,104,575. Hence, it is highly recommended for the city to spray in the areas identified from July-September.

First and foremost, the Chicago Department of Public Health (CDPH) should ensure that they are making the most of what they have. Recent reports indicate that the operators contracted by the Chicago Department of Public Health to provide mosquito spraying servicing are using an adulticide by the name of Zenivex<sup>4</sup>. Zenivex is a liquid that is intended to be dispersed through misting, killing mosquitoes upon contact, and one of the instructions for its usage is that it is not to be used at wind speeds above 10 miles per hour<sup>5</sup>. 

However, on 08/08/2013 there were 834 spraying locations that charted (based on the closest weather station) a wind speed of 10 miles per hour or greater. 

Although this may seem like a minor issue, proper collection of data requires proper application of tools, and requiring contractors to perform better and use instruments and pesticides correctly is a financially harmless way to increase efficiency.

### Spray Calculatively

As mentioned earlier, CDPH can use the data that we have analysed and provided above to spray appropriately. Currently, we have no clue how spraying is being conducted, but it doesn't seem to be according to general numbers of mosquitoes, WNV numbers of mosquitoes, or WNV rate of mosquitoes. It appears to be largely at random,, perhaps maybe even based on civilian compaints, and while there are some good hits, there are also plenty of misses. If the budget constrains spraying to only be done 7-8 times a year, more thought should be applied to it.

To start off with, we would suggest that CDPH target spraying based WNV rate of mosquitoes as a baseline, but they could also consider the characteristics that these locations have in common and work off of that to possibly find more effective locations to spray in.

### Research Alternate Insecticides

- Adulticides
    - Zenivex
        - An insecticide made from synthetic pyrethroids<sup>11</sup>, although it is able to kill mosquitoes effectively, mosquitoes have been discovered to be starting to build up a resistance to pyrethroids<sup>9 10</sup>. As a result, it may not be as effective as it once was
    - Malathion
        - While slightly more toxic than pyrethroid based insecticides<sup>12</sup>, malathion is almost much more deadly, with research showing it has a mortality rate of up to 50% with a 15 minute exposure time on mosquitoes, with permethrin, a pyrethroid insecticide, having a highest mortality rate of 8% with a 30 minute exposure time<sup>9</sup>
- Larvicides
     - Although not evident in the is dataset, CDPH apparently uses larvicides in conjunction with adulticides to control the vector population<sup>4</sup>. While known to be effective<sup>6</sup> especially if timed correctly (during the nesting weeks of the reservoir species), large-scale is more impractical and hence likely to be more costly than adulticide

### Alternate Insecticide Dispersal Methods

- Aerial Spraying
    - Aerial spraying could potentially be a more effective method at dispersing adulticides, with research showing a decrease of 57.5% in WNV-competent vector mosquitoes using a pyrethroid class insecticide<sup>2</sup>. However, health risks from blanketing an area in insecticide are a concern, and it also could be more costly overall
- Raising Community Awareness/Implementing Community Programs
    - A simple and cheap method of reducing WNV presence in the city would be to raise community awareness on how the citizens themselves can protect themselves from WNV. Some ideas could be to advise them on how they can keep their property clean and clear of any potential vector breeding grounds e.g. pooling water in pots, piles of leaves. While there is no guarantee of effectiveness, the financial cost would be quite minimal
    - Another method would be to distribute easy-to-use larvicides BTI pellets or Briquets with instructions on their use. While this would be slightly costlier, there is a higher chance of it being more effective <sup>7</sup>

### Improve Data Collection

Ultimately, our conclusion and recommendations is only as good as the data that we have been provided with. And as it stands, the data that we have been provided with is flawed in two major areas
    
   - Inconsistency in the collection date of the traps
        - An increase in the trap collection date consistency would allow us to make better predictions on the likelihood of WNV at any point in time, as well as allow us to observe timewise trends, enabling us to further identify features that affect the presence of WNV
    
   - Inconsistency in the location of the traps
        - An increase in the trap location consistency would once again allow us to make better predictions on the likelihood of WNV at any of the locations, as well as allow us to observe location-related trends, enabling us to further identify environmental features that affect the presence of WNV
    
The inconsistency could simply be due to a lack of manpower or budget. What could solve this would be to implement an automated system that would be able to detect and identify mosquitoes<sup>8</sup>, thus reducing the need for manually checking the traps. However, this technology is unfortunately currently not fully-developed yet, so while it is able to distinguish mosquitoes from different genus such as an Aedes mosquito and a Culex mosquito, it may not yet be able to distinguish mosquitoes from the same genus, which is critical to our data.
